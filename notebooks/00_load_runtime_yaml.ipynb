# Load runtime.yaml


```python
from pathlib import Path

CONFIG_PATH = Path('/Workspace/ORACLE_AIDP_STREAMING/config/runtime.yaml')
if not CONFIG_PATH.is_file():
    raise FileNotFoundError(f'Upload runtime.yaml to {CONFIG_PATH}')

config = {}
for raw_line in CONFIG_PATH.read_text(encoding='utf-8').splitlines():
    line = raw_line.strip()
    if not line or line.startswith('#') or ':' not in line:
        continue
    key, value = line.split(':', 1)
    config[key.strip()] = value.strip().strip(chr(34)).strip(chr(39))

required = ('OCI_STREAM_ENDPOINT', 'OCI_STREAM_OCID', 'OCI_CONFIG_FILE')
missing = [key for key in required if not config.get(key)]
if missing:
    raise ValueError('runtime.yaml is missing: ' + ', '.join(missing))
print('runtime.yaml loaded successfully')
print('Use config["OCI_STREAM_OCID"] and other config values in the following notebook cells.')
```
